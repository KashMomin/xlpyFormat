```python
#Function to fill expected inbound pallets
def palletFill(row, runningtotal):
    if row['PLAN RECEIPT DATE'] == 'TODAY':
        #if today is a weekend, 0, else pAvg
        if datetime.date.today().weekday() >= 5:
            return 0
        else:
            return pAvg    
    elif row['PLAN RECEIPT DATE'] == 'NEXT 10 DAYS':
        #sum up next 10 days of pAvg that aren't weekends
        for i in range(1,11):
            if (datetime.date.today()+datetime.timedelta(i)).weekday() < 5:
                runningtotal += pAvg
        return runningtotal
    #11+ days always 0
    elif row['PLAN RECEIPT DATE'] == '11+ DAYS':
        return 0
    else:
        #0 if weekend, else pAvg
        if row['PLAN RECEIPT DATE'].weekday() >= 5:
            return 0
        else:
            return pAvg
```

```python
#Loop through each labor rate
for sumType in ['DEMO', 'GOAL', 'BUDGET']:
    #################### OUTBOUND START ######################
    temp = combined.copy()
    #Calculate necessary additional columns
    temp['% CASES PICKED'] = temp.apply(lambda row: 1.0 if row['CASE PICKS'] == 0 else round(1-(row['CASE PICKS LEFT']/row['CASE PICKS']),4), axis = 1)
    temp['% PALLETS PICKED'] = temp.apply(lambda row: 1.0 if row['PALLET PICKS'] == 0 else round(1-(row['PALLET PICKS LEFT']/row['PALLET PICKS']),4), axis = 1)
    temp['CASE RATE PER HOUR'] = rateDict[sumType]['CASE_RATE']
    temp['PALLET RATE PER HOUR'] = rateDict[sumType]['PALLET_RATE']
    temp['HOURS PER PERSON'] = rateDict[sumType]['hoursPerPerson']
    temp['CASE PICK HOURS LEFT'] = temp.apply(lambda row: round(row['CASE PICKS LEFT']/row['CASE RATE PER HOUR'],4), axis = 1)
    temp['PALLET PICK HOURS LEFT'] = temp.apply(lambda row: round(row['PALLET PICKS LEFT']/row['PALLET RATE PER HOUR'],4), axis = 1)
    temp['CASE PICK MAN DAYS LEFT'] = temp.apply(lambda row: round(row['CASE PICK HOURS LEFT']/row['HOURS PER PERSON'],4), axis = 1)
    temp['PALLET PICK MAN DAYS LEFT'] = temp.apply(lambda row: round(row['PALLET PICK HOURS LEFT']/row['HOURS PER PERSON'],4), axis = 1)
    temp['DUMMY1'] = '  '
    temp['DUMMY2'] = '  '
    #reorder columns
    temp = temp[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'ORDERS', 'LINES', 'CASES', 'DUMMY1', 'CASE PICKS', '% CASES PICKED', 'CASE PICKS LEFT', 'EST PALLETS FROM CASE PICKS LEFT', 'CASE RATE PER HOUR', 'CASE PICK HOURS LEFT', 'HOURS PER PERSON', 'CASE PICK MAN DAYS LEFT', 'DUMMY2', 'CASES ON PALLET PICKS', 'PALLET PICKS', '% PALLETS PICKED', 'PALLET PICKS LEFT', 'PALLET RATE PER HOUR', 'PALLET PICK HOURS LEFT', 'HOURS PER PERSON', 'PALLET PICK MAN DAYS LEFT']]
    
    tempShrink = temp[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'EST PALLETS FROM CASE PICKS LEFT']].rename(columns = {'EST PALLETS FROM CASE PICKS LEFT':'PALLETS TO SHRINK WRAP'}).copy()
    tempShrink['SHRINK WRAP RATE PPH'] = rateDict[sumType]['SHRINK_RATE']
    tempShrink['SHRINK WRAP HOURS LEFT'] = tempShrink.apply(lambda row: round(row['PALLETS TO SHRINK WRAP']/row['SHRINK WRAP RATE PPH'],4), axis = 1)
    tempShrink['SHRINK WRAP MAN DAYS LEFT'] = tempShrink.apply(lambda row: round(row['SHRINK WRAP HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    tempStage = temp[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'EST PALLETS FROM CASE PICKS LEFT', 'PALLET PICKS LEFT']].copy()
    tempStage['PALLETS TO STAGE'] = tempStage['EST PALLETS FROM CASE PICKS LEFT'] + tempStage['PALLET PICKS LEFT']
    tempStage = tempStage.drop(['EST PALLETS FROM CASE PICKS LEFT', 'PALLET PICKS LEFT'], axis = 1)
    tempStage['STAGE RATE PPH'] = rateDict[sumType]['STAGE_RATE']
    tempStage['STAGE HOURS LEFT'] = tempStage.apply(lambda row: round(row['PALLETS TO STAGE']/row['STAGE RATE PPH'],4), axis = 1)
    tempStage['STAGE MAN DAYS LEFT'] = tempStage.apply(lambda row: round(row['STAGE HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    tempLoad = temp[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'EST PALLETS FROM CASE PICKS LEFT', 'PALLET PICKS LEFT']].copy()
    tempLoad['PALLETS TO LOAD AND SHIP'] = tempLoad['EST PALLETS FROM CASE PICKS LEFT'] + tempLoad['PALLET PICKS LEFT']
    tempLoad = tempLoad.drop(['EST PALLETS FROM CASE PICKS LEFT', 'PALLET PICKS LEFT'], axis = 1)
    tempLoad['LOAD AND SHIP RATE PPH'] = rateDict[sumType]['LOAD_RATE']
    tempLoad['LOAD AND SHIP HOURS LEFT'] = tempLoad.apply(lambda row: round(row['PALLETS TO LOAD AND SHIP']/row['LOAD AND SHIP RATE PPH'],4), axis = 1)
    tempLoad['LOAD AND SHIP MAN DAYS LEFT'] = tempLoad.apply(lambda row: round(row['LOAD AND SHIP HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    tempReplen = temp[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'EST PALLETS FROM CASE PICKS LEFT']].rename(columns = {'EST PALLETS FROM CASE PICKS LEFT':'PALLETS TO REPLENISH'}).copy()
    tempReplen['REPLENISHMENT RATE PPH'] = rateDict[sumType]['REPLEN_RATE']
    tempReplen['REPLENISHMENT HOURS LEFT'] = tempReplen.apply(lambda row: round(row['PALLETS TO REPLENISH']/row['REPLENISHMENT RATE PPH'],4), axis = 1)
    tempReplen['REPLENISHMENT MAN DAYS LEFT'] = tempReplen.apply(lambda row: round(row['REPLENISHMENT HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    toMerge = [temp[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'CASE PICK HOURS LEFT', 'CASE PICK MAN DAYS LEFT', 'PALLET PICK HOURS LEFT', 'PALLET PICK MAN DAYS LEFT']],
               tempShrink[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'SHRINK WRAP HOURS LEFT', 'SHRINK WRAP MAN DAYS LEFT']],
               tempStage[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'STAGE HOURS LEFT', 'STAGE MAN DAYS LEFT']],
               tempLoad[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'LOAD AND SHIP HOURS LEFT', 'LOAD AND SHIP MAN DAYS LEFT']],
               tempReplen[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'REPLENISHMENT HOURS LEFT', 'REPLENISHMENT MAN DAYS LEFT']]
              ]
    
    #merge all frames together
    otherSummary = functools.reduce(lambda  left,right: pd.merge(left,right,on=['PLAN SHIP DATE', 'PLAN SHIP DOW'], how='left'), toMerge).fillna(0)
    
    #otherSummary['TOTAL OTHER LABOR HOURS LEFT'] = otherSummary.apply(lambda row: row['SHRINK WRAP HOURS LEFT'] + row['STAGE HOURS LEFT'] + row['LOAD AND SHIP HOURS LEFT'] + row['REPLENISHMENT HOURS LEFT'], axis = 1)
    otherSummary['TOTAL PICKING LABOR HOURS LEFT'] = otherSummary.apply(lambda row: row['CASE PICK HOURS LEFT'] + row['PALLET PICK HOURS LEFT'], axis = 1)
    otherSummary['TOTAL OB LABOR HOURS LEFT'] = otherSummary.apply(lambda row: row['TOTAL OTHER LABOR HOURS LEFT'] + row['TOTAL PICKING LABOR HOURS LEFT'], axis = 1)
    otherSummary['TOTAL OB MAN DAYS LEFT'] = otherSummary.apply(lambda row: round(row['TOTAL OB LABOR HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    otherSummary = otherSummary[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'TOTAL OTHER LABOR HOURS LEFT', 'TOTAL PICKING LABOR HOURS LEFT', 'TOTAL OB LABOR HOURS LEFT', 'TOTAL OB MAN DAYS LEFT']]
    
    toMergeOverview.append(otherSummary[['PLAN SHIP DATE', 'PLAN SHIP DOW', 'TOTAL OB LABOR HOURS LEFT', 'TOTAL OB MAN DAYS LEFT']].rename(columns = {'TOTAL OB LABOR HOURS LEFT':'TOTAL OB LABOR HOURS ' + sumType, 'TOTAL OB MAN DAYS LEFT':'TOTAL OB MAN DAYS ' + sumType}))
    
    #################### INBOUND START ######################
    
    tempReceiptList = combined[['PLAN SHIP DATE', 'PLAN SHIP DOW']].rename(columns = {'PLAN SHIP DATE':'PLAN RECEIPT DATE', 'PLAN SHIP DOW':'PLAN RECEIPT DOW'}).copy()
    tempReceiptList['PALLETS EXPECTED'] = tempReceiptList.apply(palletFill, axis = 1, runningtotal = 0)
    tempReceiptList['SINGLE SKU PALLETS'] = tempReceiptList.apply(lambda row: round(row['PALLETS EXPECTED']*.95, 0), axis = 1)
    tempReceiptList['MIXED SKU PALLETS'] = tempReceiptList.apply(lambda row: round(row['PALLETS EXPECTED']*.05, 0), axis = 1)
    tempReceiptList['TOTAL PALLETS'] = tempReceiptList.apply(lambda row: row['SINGLE SKU PALLETS'] + (row['MIXED SKU PALLETS']*4), axis = 1)
    
    tempClerkFixed = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'TOTAL PALLETS']].copy()
    tempClerkFixed['HOURS'] = tempClerkFixed.apply(lambda row: 0 if row['TOTAL PALLETS'] == 0 else 8, axis = 1)
    tempClerkFixed['SHIFT'] = tempClerkFixed.apply(lambda row: 0 if row['TOTAL PALLETS'] == 0 else 8, axis = 1)
    tempClerkFixed['CLERK MAN DAYS LEFT'] = tempClerkFixed.apply(lambda row: 0 if row['TOTAL PALLETS'] == 0 else 1, axis = 1)
    del tempClerkFixed['TOTAL PALLETS']
    
    tempOffload = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'PALLETS EXPECTED']].rename(columns = {'PALLETS EXPECTED':'PALLETS TO OFFLOAD'}).copy()
    tempOffload['OFFLOAD RATE PPH'] = rateDict[sumType]['OFFLOAD_RATE']
    tempOffload['OFFLOAD HOURS LEFT'] = tempOffload.apply(lambda row: round(row['PALLETS TO OFFLOAD']/row['OFFLOAD RATE PPH'],4), axis = 1)
    tempOffload['OFFLOAD MAN DAYS LEFT'] = tempOffload.apply(lambda row: round(row['OFFLOAD HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    tempSegPallet = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'MIXED SKU PALLETS']].copy()
    tempSegPallet['CASES TO SEGREGATE'] = tempSegPallet.apply(lambda row: 143*row['MIXED SKU PALLETS'], axis = 1)
    tempSegPallet['SEGREGATE RATE CPH'] = rateDict[sumType]['SEGREGATE_RATE']
    tempSegPallet['SEGREGATE HOURS LEFT'] = tempSegPallet.apply(lambda row: round(row['CASES TO SEGREGATE']/row['SEGREGATE RATE CPH'],4), axis = 1)
    tempSegPallet['SEGREGATE MAN DAYS LEFT'] = tempSegPallet.apply(lambda row: round(row['SEGREGATE HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    del tempSegPallet['MIXED SKU PALLETS']
    
    tempValPallet = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'TOTAL PALLETS']].rename(columns = {'TOTAL PALLETS':'PALLETS TO VALIDATE'}).copy()
    tempValPallet['VALIDATE RATE PPH'] = rateDict[sumType]['VALIDATE_RATE']
    tempValPallet['VALIDATE HOURS LEFT'] = tempValPallet.apply(lambda row: round(row['PALLETS TO VALIDATE']/row['VALIDATE RATE PPH'],4), axis = 1)
    tempValPallet['VALIDATE MAN DAYS LEFT'] = tempValPallet.apply(lambda row: round(row['VALIDATE HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    tempNCIPallet = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'TOTAL PALLETS']].copy()
    tempNCIPallet['EXPECTED NCI PALLETS'] = tempNCIPallet.apply(lambda row: math.ceil(.05*row['TOTAL PALLETS']), axis = 1)
    tempNCIPallet['NCI RATE PPH'] = rateDict[sumType]['NCI_RATE']
    tempNCIPallet['NCI HOURS LEFT'] = tempNCIPallet.apply(lambda row: round(row['EXPECTED NCI PALLETS']/row['NCI RATE PPH'],4), axis = 1)
    tempNCIPallet['NCI MAN DAYS LEFT'] = tempNCIPallet.apply(lambda row: round(row['NCI HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    del tempNCIPallet['TOTAL PALLETS']
    
    tempLabel = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'TOTAL PALLETS']].rename(columns = {'TOTAL PALLETS':'PALLETS TO LABEL'}).copy()
    tempLabel['LABEL RATE PPH'] = rateDict[sumType]['LABEL_RATE']
    tempLabel['LABEL HOURS LEFT'] = tempLabel.apply(lambda row: round(row['PALLETS TO LABEL']/row['LABEL RATE PPH'],4), axis = 1)
    tempLabel['LABEL MAN DAYS LEFT'] = tempLabel.apply(lambda row: round(row['LABEL HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    tempPutaway = tempReceiptList[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'TOTAL PALLETS']].rename(columns = {'TOTAL PALLETS':'PALLETS TO PUTAWAY'}).copy()
    tempPutaway['PUTAWAY RATE PPH'] = rateDict[sumType]['PUTAWAY_RATE']
    tempPutaway['PUTAWAY HOURS LEFT'] = tempPutaway.apply(lambda row: round(row['PALLETS TO PUTAWAY']/row['PUTAWAY RATE PPH'],4), axis = 1)
    tempPutaway['PUTAWAY MAN DAYS LEFT'] = tempPutaway.apply(lambda row: round(row['PUTAWAY HOURS LEFT']/rateDict[sumType]['hoursPerPerson'],4), axis = 1)
    
    toMergeInbound = [tempClerkFixed[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'CLERK MAN DAYS LEFT']],
               tempOffload[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'OFFLOAD MAN DAYS LEFT']],
               tempSegPallet[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'SEGREGATE MAN DAYS LEFT']],
               tempValPallet[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'VALIDATE MAN DAYS LEFT']],
               tempNCIPallet[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'NCI MAN DAYS LEFT']],
               tempLabel[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'LABEL MAN DAYS LEFT']],
               tempPutaway[['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW', 'PUTAWAY MAN DAYS LEFT']]
              ]
    
    #merge all frames together
    inboundSummary = functools.reduce(lambda  left,right: pd.merge(left,right,on=['PLAN RECEIPT DATE', 'PLAN RECEIPT DOW'], how='left'), toMergeInbound).fillna(0)
    inboundSummary['TOTAL MAN DAYS LEFT'] = inboundSummary.apply(lambda row: row['CLERK MAN DAYS LEFT'] + row['OFFLOAD MAN DAYS LEFT'] + row['SEGREGATE MAN DAYS LEFT'] + row['VALIDATE MAN DAYS LEFT'] + row['NCI MAN DAYS LEFT'] + row['LABEL MAN DAYS LEFT'] + row['PUTAWAY MAN DAYS LEFT'], axis = 1)
    
    outputDict[sumType] = [temp, tempShrink, tempStage, tempLoad, tempReplen, otherSummary, tempReceiptList, tempClerkFixed, tempOffload, tempSegPallet, tempValPallet, tempNCIPallet, tempLabel, tempPutaway, inboundSummary]
```

```python
#function to adjust column width, taking inputs of df and sheet name    
def adjust_cols_width(width_list,sheet_name):

    for i in range(0,len(width_list)):
        writer.sheets[sheet_name].set_column(i, i, width_list[i])
```

```python
def FormatHeaders(df, sheet_name, startrow, startcol):
    
    # Write the column headers with the defined format.
    for col_num, value in enumerate(df.columns.values):
        if 'DUMMY' in value:
            writer.sheets[sheet_name].write(startrow, startcol + col_num, '', header_format)
        else:
            writer.sheets[sheet_name].write(startrow, startcol + col_num, value, header_format)
```

```python
def FormatCells(sheet_name, fType, rowRange, colList):
    for c in colList:
        rangeStr = c + str(rowRange[0]) + ':' + c + str(rowRange[1])
        writer.sheets[sheet_name].conditional_format(rangeStr, {'type': 'no_errors','format': fType})
```
